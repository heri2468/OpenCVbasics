```
import cv2
from skimage import io
from pylab import *
```

```
image = cv2.imread('shubhu.jpeg')
io.imshow(image)
```

```
_, img1 = cv2.threshold(image, 50, 255,0 )
_, img2 = cv2.threshold(image, 50, 255,1 )
_, img3 = cv2.threshold(image, 50, 255,2 )
_, img4 = cv2.threshold(image, 50, 255,3 )
_, img5 = cv2.threshold(image, 50, 255,4 )
```

```
figure(1)
io.imshow(img1)
figure(2)
io.imshow(img2)
figure(3)
io.imshow(img3)
figure(4)
io.imshow(img4)
figure(5)
io.imshow(img5)


```
