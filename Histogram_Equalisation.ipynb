```
import cv2
from skimage import io
from pylab import *
```

```
image = cv2.imread('shubhu.jpeg')
io.imshow(image)
```

```
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
#Apply equalize histogram
src_eqlzd = cv2.equalizeHist(image) #Performs Histogram Equalization

```

```
#Show both images
figure(0)
io.imshow(image)
figure(1)
io.imshow(src_eqlzd)
```
