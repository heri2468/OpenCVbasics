```python
import cv2
from skimage import io
```

```python
#blending two images
#reading image 1
img1 = cv2.imread(r'C:\Users\ritesh\Desktop\new.jpg')
img2 = cv2.imread(r'C:\Users\ritesh\Desktop\newrit.jpg')
```

```python
alpha = 0.60
beta = 0.40
#blend
final_image = cv2.addWeighted(img1,alpha,img2,beta,0.0)
```

```python
io.imshow(final_image)
```

```python
cv2.imshow("img",final_image)
while(True):
    k = cv2.waitKey(33)
    if k == -1:  # if no key was pressed, -1 is returned
        continue
    else:
        break
cv2.destroyWindow('img')
```

```python
io.imsave("blended.jpg",final_image)
```
